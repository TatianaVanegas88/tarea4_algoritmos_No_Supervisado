```python
# Importar librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

# Cargar el dataset
data = pd.read_csv("Mall_Customers.csv")

# Mostrar las primeras filas y la información del dataset
print(data.head())
print(data.info())
```

```python
# Resumen estadístico
print(data.describe())

# Visualización de la distribución de ingresos anuales
sns.histplot(data['Annual Income (k$)'], kde=True)
plt.title('Distribución de Ingresos Anuales')
plt.show()

# Relación entre ingreso anual y puntaje de gasto
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Gender', data=data)
plt.title('Relación entre Ingreso Anual y Puntaje de Gasto')
plt.show()

# Detección de valores atípicos en ingresos anuales
sns.boxplot(data['Annual Income (k$)'])
plt.title('Outliers en Ingreso Anual')
plt.show()
```
